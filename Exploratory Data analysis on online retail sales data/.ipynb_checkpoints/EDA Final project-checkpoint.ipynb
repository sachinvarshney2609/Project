# **6. Data Pre-Processing**
---

- This section is emphasised on performing data manipulation over unstructured data for further processing and analysis.

- To modify unstructured data to strucuted data you need to verify and manipulate the integrity of the data by:
  - Handling missing data,

  - Handling redundant data,

  - Handling inconsistent data,

  - Handling outliers,

  - Handling typos

Check for Null values

```python
df_online.isnull().sum()
```

Removing Missing/NaN Data from dataset

```python
df_online[df_online['Postal Code'].isnull()].head()
```

 Postal code is not significant and contains more than 85% null values , so it can be dropped.

```python
df_online.drop(['Postal Code'],axis= 1,inplace=True)
```

Adding a new column unit price

```python
df_online['Unit Price'] = df_online['Sales'] / df_online['Quantity']
df_online.head()
```

Coverting Order Date and Shipdate to datetime format

```python
df_online['Order Date'] = pd.to_datetime(df_online['Order Date'],format='%d-%m-%Y')
df_online['Ship Date'] = pd.to_datetime(df_online['Ship Date'],format='%d-%m-%Y')
```

```python
df_online.info()
```

Converted Order Date and Ship Date to datetime obejct to carry only Year

## 8.1. Univariate Analysis

### 8.1.1. Top 5 Sold products.

###  Which are top 5 product sold?

```python
df_top5_products = df_online.groupby('Product Name')['Quantity'].sum()
df_top5_products.nlargest(5,keep='all').plot(kind='bar', figsize=(15,6), color='orange', fontsize=13, yticks=np.arange(0, 1000, 100))
plt.savefig('top 5 sold products.png')
plt.show()
```

Eldon File Cart, Single Width <br>

<a name = Section812></a>
### 8.1.2. Top 5 revenue generating product

### What are top 5 revenue generating products ?

```python
df_top5_products = df_online.groupby('Product Name')['Sales'].sum()
df_top5_products.nlargest(5,keep='all').plot(kind='bar', figsize=(10,5), color='blue', fontsize=9, yticks=np.arange(0, 100000, 25000))
plt.savefig('top 5 products revenue.png')
plt.show()
```

<a name = Section813></a>
### 8.1.3. Average shipping time

### What is average shipping time for sold products?

```python
#df_online.drop(['Shipping time','Shipping days1'],axis=1,inplace=True)


df_online['Shipping days'] = pd.to_numeric((df_online['Ship Date'] - df_online['Order Date'])/10e13)

df_online['Shipping days'].value_counts().plot(kind='bar', figsize=(12, 7), color='green')

plt.xlabel('Shipping time(in days)')
plt.ylabel('Frequency')
plt.title('Bar chart showing the shipping time (in days) for order placed')

plt.show()



```

Majority of order are shipped within 3 to 4 days and none of them are delivered after 7 days


<a name = Section814></a>
### 8.1.4. Unit Price Trends

### What is unit price ranges for order placed?

```python
#df_online.drop(['Shipping time','Shipping days1'],axis=1,inplace=True)

bins = np.arange(0,300,10)
df_online['Unit Price'].plot(kind='hist', bins = bins, figsize=(12, 7), color='brown', xticks=bins)

plt.xlabel('Unit Price')
plt.ylabel('Frequency')
plt.title('Histogram showing the unit price ranges  for order placed')

plt.show()
```

Majority of the products are in the range of 0-60 USD.
